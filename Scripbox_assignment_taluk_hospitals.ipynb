#### list_points corresponds to hospitals location coordinates. Following step is to convert the data into a list of points

```python
list_points=[]
x=''
y=''
count=0
for i in range(len(data1)):
    count=count+1
    try:
        x=data1[i].split(',')[0]
        x_f=float(x)
        y=data1[i].split(',')[1]
        y_f=float(y)
        #print(x_f,y_f)
        list_points.append([y_f,x_f])
    except (KeyError,ValueError):
        pass
        
```

### Calculating total number of hospitals present in each Taluk

```python
fin_1=[]
fin_2=[]
fin_3=[]
fin_4=[]
fin_5=[]
fin_6=[]
for feature in data3['features']:
    polygon = shape(feature['geometry'])
    count_=0
    
    for j in tqdm(range(len(list_points))):
        if polygon.contains(Point(list_points[j])):
            count_=count_+1
        else:
            pass
    
    fin_1.append(feature['properties']['NAME_0'])
    fin_3.append(feature['properties']['NAME_1'])
    fin_2.append(count_)
    fin_4.append(area(feature['geometry']))
    fin_5.append(feature['properties']['NAME_3'])
    fin_6.append(feature['properties']['ID_3'])
```

### Converting to a dataframe

```python
df_taluk_hosp=pd.DataFrame(columns=['State','Taluk','ID','Area','Country','Number of Hospitals'])
df_taluk_hosp['Country']=fin_1
df_taluk_hosp['State']=fin_3
df_taluk_hosp['Area']=fin_4
df_taluk_hosp['Taluk']=fin_5
df_taluk_hosp['ID']=fin_6
df_taluk_hosp['Number of Hospitals']=fin_2
df_taluk_hosp.sort_values(by='Number of Hospitals',ascending = False).head()
```

### Funtion to calculate the shortest distance between a polygon and a point

```python
def cal_dist(x,y):
    
    return x.exterior.distance(y)
```
